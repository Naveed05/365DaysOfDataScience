```python
import matplotlib.pyplot as plt  # for visualization
import seaborn as sns            # statistical visualization
import pandas as pd
import warnings

warnings.filterwarnings('ignore')  # to ignore warnings

# Load dataset
df = pd.read_csv(r"C:\Users\mirza\Desktop\datasets\data.csv")

# Set figure size globally
plt.rcParams['figure.figsize'] = (6, 2)

# Show columns (optional step)
# print(df.columns)

# Plot with histogram + KDE curve
vis1 = sns.displot(df["InternetUsers"], kde=True)
plt.show()
```

```python
vis2 = sns.distplot(df["InternetUsers"], bins = 15 )
plt.show(vis2)

#<seaborn.axisgrid.FacetGrid at 0x1dfdb3ee000> this putput


```

```python
vis3 = sns.boxplot(df["InternetUsers"])
plt.show(vis3)
```

```python
vis4 = sns.barplot(df["InternetUsers"])
plt.show(vis4)
```

```python
vis5 = sns.boxplot(data = df, x ='IncomeGroup', y= 'BirthRate')
plt.show()
```

```python
vis5 = sns.lmplot(data = df, x ='InternetUsers', y= 'BirthRate', fit_reg=False)
plt.show(vis5)
```
